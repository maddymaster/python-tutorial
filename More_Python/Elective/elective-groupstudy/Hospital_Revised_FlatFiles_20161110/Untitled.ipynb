```python
# summing up the missing values (column-wise)
round(100*(hospitals_infections.isnull().sum()/len(hospitals_infections.index)), 2)
```

```python
#dropping columns with more than 50% missing data hospitals_infections

hospitals_infections = hospitals_infections.drop('Compared to National', axis=1)
hospitals_infections = hospitals_infections.drop('Footnote', axis=1)


round(100*(hospitals_infections.isnull().sum()/len(hospitals_infections.index)), 2)
```

```python
#outpatient_imaging
round(100*(outpatient_imaging.isnull().sum()/len(outpatient_imaging.index)), 2)
```

```python
#dropping missing values outpatient_imaging

outpatient_imaging = outpatient_imaging.drop('Footnote', axis=1)


round(100*(outpatient_imaging.isnull().sum()/len(outpatient_imaging.index)), 2)
```

```python
#readmission
round(100*(readmission.isnull().sum()/len(readmission.index)), 2)
```

```python
#dropping missing values readmission

readmission = readmission.drop('Footnote', axis=1)


round(100*(readmission.isnull().sum()/len(readmission.index)), 2)
```

```python
#complications
round(100*(complications.isnull().sum()/len(complications.index)), 2)
```

```python
#removing missing values for complications

complications = complications.drop('Footnote', axis=1)


round(100*(complications.isnull().sum()/len(complications.index)), 2)
```

```python
#timely_effective_care
round(100*(timely_effective_care.isnull().sum()/len(timely_effective_care.index)), 2)
```

```python
#removing missing values for timely_effective_care

timely_effective_care = timely_effective_care.drop('Footnote', axis=1)


round(100*(timely_effective_care.isnull().sum()/len(timely_effective_care.index)), 2)
```

```python
#hcaps
round(100*(hcaps.isnull().sum()/len(hcaps.index)), 2)
```

```python
#removing missing values for hcaps

hcaps = hcaps.drop('Patient Survey Star Rating Footnote', axis=1)
hcaps = hcaps.drop('HCAHPS Answer Percent Footnote', axis=1)
hcaps = hcaps.drop('Number of Completed Surveys Footnote', axis=1)
hcaps = hcaps.drop('Survey Response Rate Percent Footnote', axis=1)


round(100*(hcaps.isnull().sum()/len(hcaps.index)), 2)
```
