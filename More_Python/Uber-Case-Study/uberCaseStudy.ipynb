Notice that there are columns have almost 40% and 60% missing values in Driver id and Timestamp respectively

```python
# For now I will remove the drop timestamp column.
# removing the drop timestamp column
df = df.drop('Drop timestamp', axis=1)
```

```python
#checking again
round(100*(df.isnull().sum()/len(df.index)), 2)
```

```python
#treating missing values in Driverid
df[df.isnull().sum(axis=1) < 5]
```

```python
# count the number of rows having > 2 missing values
# use len(df.index)
len(df[df.isnull().sum(axis=1) > 2].index)
```

```python
#checking summary again
round(100*(df.isnull().sum()/len(df.index)), 2)
```

```python
#checking distribution
df['Driver id'].describe()
```

```python
# rows having Driver Id missing
df[np.isnan(df['Driver id'])]
```

```python
df.loc[np.isnan(df['Driver id']), ['Driver id']] = df['Driver id'].mean()
```

```python
round(100*(df.isnull().sum()/len(df.index)), 2)
```
